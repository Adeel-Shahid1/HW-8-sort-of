## Q1.

1.

Simple Linear Regression analyzes the relationship between one predictor and one outcome variable, while Multiple Linear Regression allows for multiple predictor vairables.
(y= b0 + b1x + ϵ vs y = b0 + b1x1 + b2x2 + ... + bpxp + ε)

2.

A continuous variable allows for any value within a range to be used.In the context of simple linear regression it looks like this:
#### outcome = b0 + b1(continuous) 

An indicator/dummy variable represents categorical data. It takes on the value of 1 if a condition is met and 0 if it is not.In the context of simple linear regression it looks like this:
#### outcome=b0 + β1 * 1(category)

1(category) is 1 for one group and 0 for the other.

3.

Adding an indicator variable to a continuous variable model transforms it into a Multiple Linear Regression which is represented by two parallel lines for each level of the indicator variable.

#### Simple Linear Regression with a continuous variable:
outcome= b0 + b1 * continuous predictor

#### Multiple Linear Regression (MLR) with a continuous and an indicator variable:
outcome= b0 + b1 * ⋅continuous predictor + b2 * 1(category)

The model can now account for variations in the dependent variable based on group/category.

4.

Adding an interaction term allows the effect of the continuous variable to differ based on the category represented by the indicator variable.

outcome = β₀ + β₁ * continuous_predictor + β₂ * 1(indicator_variable) + β₃ * continuous_predictor * 1(indicator_variable)

This model can show how the relationship between the continuous variable and the dependent variable changes depending on the category

5.

When using only indicator variables from a non-binary categorical variable, the Multiple Linear Regression model creates multiple parallel lines for each category level. A categorical variable with K levels uses K-1 indicator variables (to avoid multicollinearity). β₀ serves as a baseline while other indicator variables show mean differences between other groups and the baseline.
This can be understood with this provided example:

A categorical variable "color" with levels: "red", "blue", and "green" can be encoded with:

    1(color="blue"): 1 if blue, 0 otherwise.
    1(color="green"): 1 if green, 0 otherwise.
    
The linear form is:
outcome = β₀ + β₁ * 1(color="blue") + β₂ * 1(color="green")
Here:
"red" is the baseline.
β₀ is the mean of the "red" group.
β₁ is the mean difference between "blue" and "red" groups.
β₂ is the mean difference between "green" and "red" groups.

Outcome Variable: The effectiveness of the advertising campaigns, which could be measured in terms of sales, customer engagement, or brand awareness.

Predictor Variables:

TV Advertising Spend (Continuous Variable): The amount spent on TV ads.
Online Advertising Spend (Continuous Variable): The amount spent on online ads.

It might be important to take into account the interaction between the two advertising since the effectiveness of the TV ad might depend on the amount spent on online advertising and vice versa, there is a potential interaction effect between these two advertising mediums. 

### Without Interaction:

Equation: 

Effectiveness = B0 + B1 (TV Spend)+ B2(Online Spend)
This model assumes that the effects of TV and online advertising are additive and independent.

Model assumes that the effects of TV and online advertising are additive and independent.
You would predict effectiveness based solely on the individual contributions of TV and online spending. The prediction assumes that increasing one type of spending does not affect the impact of the other.

### With Interaction:

Equation: 
Effectiveness = B0 + B1 (TV Spend)+ B2 (Online Spend)+ B12(TV Spend×Online Spend)

In this model, the interaction term B12(TV Spend×Online Spend) captures how the effectiveness of one advertising medium depends on the spending of the other. For example, the impact of TV spend on effectiveness may increase or decrease depending on the level of online spend.

This model allows for a more nuanced prediction, where the effectiveness can change based on the combination of spending levels on both platforms. For example, spending more on online ads might enhance the effectiveness of TV ads if they are also increased.

Effectiveness: B0 + B1 (TV High)+ B2 (Online High)+ B12 (TV High×Online High)
